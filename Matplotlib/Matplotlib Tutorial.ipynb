# Welcome to Matplotlib Tutorial

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.plot([1,2,3], [2,5,7])
plt.show()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.plot([1,2,3], [2,5,7])
plt.title("Sandip's Data")
plt.ylabel("This is Y")
plt.xlabel("This is X")
plt.show()
```

```python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')
x = [1,5,4]
y = [28, 65, 98]
x2 = [4, 7, 9]
y2= [5, 8, 6]

plt.plot(x,y, label= 'first', linewidth=3)
plt.plot(x2,y2, label= 'second')
plt.legend()
plt.show()
```

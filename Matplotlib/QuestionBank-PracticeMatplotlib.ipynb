**Write a program to plot multiple subplots in a single figure.**


```python
import matplotlib.pyplot as plt

# Creating data
x = range(10)
y1 = [val**2 for val in x]
y2 = [val**3 for val in x]

# Creating subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
ax1.plot(x, y1, label="y = x^2")
ax1.set_title("Quadratic")
ax1.legend()

ax2.plot(x, y2, label="y = x^3", color="red")
ax2.set_title("Cubic")
ax2.legend()

plt.show()
```

**Explain in detail the process of creating a stacked bar chart in Matplotlib.**

```python
# A stacked bar chart can be created by calling plt.bar() multiple times with the bottom parameter. 
# This parameter specifies the starting point of each bar stack.

import matplotlib.pyplot as plt

categories = ['A', 'B', 'C']
values1 = [5, 7, 3]
values2 = [6, 2, 5]

plt.bar(categories, values1, label='Group 1')
plt.bar(categories, values2, bottom=values1, label='Group 2')

plt.legend()
plt.show()
```

**Write a program to create a customized scatter plot with annotations for each point.**

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Plot
plt.scatter(x, y, color='blue', marker='o')

# Annotate points
for i, txt in enumerate(y):
    plt.annotate(f'({x[i]}, {y[i]})', (x[i], y[i]))

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Annotated Scatter Plot')
plt.show()
```
